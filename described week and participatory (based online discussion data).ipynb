```python
discussion_data = pd.read_csv('C:/Users/Lee/Downloads/online-discussion-dataset-master/online-discussion-dataset-master/orig_data.txt') # load dataset from local machine
type(discussion_data) #view dataset  datatype
len(discussion_data)  # view dataset length
discussion_data.head() # view dataset first 5 lines
```

```python
week_rel_participatory = discussion_data.groupby(['week',discussion_data['participatory_role']]).size()  # grouped data with two columns, week , participatory espectively.
week_rel_participatory.head() #view data type, which is not table structure but tree structure.
week_rel_participatory = week_rel_participatory.unstack() # finished transform from tree structure to table structure.
```
